```python
station_names = []
for name in crime_anal_police['관서명']:
    station_names.append('서울' + str(name[:-1]) + '경찰서')

station_names
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_names:
    tmp = gmaps.geocode(name, language='ko')
    
    station_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc = tmp[0]['geometry']['location']
    station_lat.append(tmp_loc['lat'])
    station_lng.append(tmp_loc['lng'])
    
    print(name + ' ---> ' + tmp[0].get('formatted_address'))
```

```python
crime_anal_raw = pd.read_csv(
    '../data/02. crime_in_Seoul_include_gu_name.csv'
    , encoding='utf-8'
    , index_col = 0
)
crime_anal_raw.head()
```

```python
crime_anal = pd.pivot_table(
      data=crime_anal_raw
    , index='구별'
    , aggfunc=np.sum
)
crime_anal.head()
```

```python
del crime_anal['관서명']
crime_anal.head()
```

```python
crime = ['강간', '강도', '살인', '절도', '폭력']
for n in crime:
    crime_anal[f'{n}검거율'] = crime_anal[f'{n} 검거'] / crime_anal[f'{n} 발생'] * 100
```

```python
crime_anal.head()
```

```python
for n in crime:
    del crime_anal[f'{n} 검거']
crime_anal.head()
```

```python
con_list = crime_anal.columns.to_list()[5:]
for column in con_list:
    crime_anal.loc[crime_anal[column] > 100, column] = 100
    
crime_anal[con_list].head()
```

```python
arrestRate = crime_anal[con_list]
arrestRate.head()
```

```python
crime_anal.head()
```

```python
crime_anal.rename(
    columns={
          '강간 발생': '강간'
        , '강도 발생': '강도'
        , '살인 발생': '살인'
        , '절도 발생': '절도'
        , '폭력 발생': '폭력'
    }
    , inplace=True
)
crime_anal.columns
```

```python
crime_anal.head()
```

```python
from sklearn import preprocessing

cols = ['강간', '강도', '살인', '절도', '폭력']

x = crime_anal[cols].values

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

crime_anal_norm = pd.DataFrame(
      data=x_scaled
    , columns=cols
    , index=crime_anal.index
)
crime_anal_norm.head()
```

```python
colums = crime_anal.columns.to_list()[5:]
# print(colums)
crime_anal_norm[colums] = crime_anal[colums]
crime_anal_norm.head()
```

```python
result = pd.read_csv(
    '../data/01. CCTV_result.csv'
    , encoding='utf-8'
    , index_col='구별'
)
result.head()
```

```python
crime_anal_norm[['인구수', '소계']] = result[['인구수', '소계']]
crime_anal_norm.head()
```

```python
col = ['강간', '강도', '살인', '절도', '폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(0, 14, 100)
print(x)
```

```python
y1 = np.sin(x)
y2 = 2 * np.sin(x - 0.5)
y3 = 3 * np.sin(x - 1)
y4 = 4 * np.sin(x - 1.5)

sns.set_style('ticks')
plt.figure(figsize=(10, 6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
sns.despine(left=True, bottom=True, right=False)
plt.show()
```
