```python
# https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago
from bs4 import BeautifulSoup
from urllib.request import urlopen

url_base = 'https://www.chicagomag.com'
url_sub = '/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago'

url = url_base + url_sub

# html = urlopen(url)

# soup = BeautifulSoup(html, 'html.parser')
```

```python
import requests

request_headers = {
    'User-Agent': 'Mozilla/5.0'
}
resp = requests.get(url=url, headers=request_headers).text

print(type(resp))
soup = BeautifulSoup(resp, 'html.parser')
```

```python
ranks = []
main_menus = []
cafe_names = []
url_adds = []

list_soup = soup.find_all('div', class_='sammy')

for item in list_soup:
    ranks.append(item.find(class_='sammyRank').text)
    tmp_str = item.find(class_='sammyListing').text
    main_menus.append(tmp_str.split('\n')[0])
    cafe_names.append(tmp_str.split('\n')[1])
    tmp_url = item.find('a')['href']
    if tmp_url.startswith('https://'):
        url_adds.append(tmp_url)
    else:
        url_adds.append(url_base + tmp_url)
```

```python
import numpy as np
import pandas as pd

data = pd.DataFrame({
      'Rank': ranks
    , 'Menu': main_menus
    , 'Cafe': cafe_names
    , 'Url': url_adds
})

data.head(10)
```

```python
from tqdm import tqdm

price = []
address = []

for idx in tqdm(df.index):
    url_str = df['Url'][idx]
    
    request_headers = {
        'User-Agent': 'Mozilla/5.0'
    }
    resp = requests.get(url=url_str, headers=request_headers).text
    
    soup = BeautifulSoup(resp, 'html.parser')
    
    get_str = soup.find('p', class_='addy').get_text()
    
    price.append(get_str.split()[0][:-1])
    address.append(' '.join(get_str.split()[1:-2]))
```

```python
df.head()
```

```python
df['Price'] = price
df['Address'] = address
df.head()
```

```python
df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.head()
```

```python
df.set_index('Rank', inplace=True)
df.head()
```

```python
df = pd.read_csv(
      '../data/03. best_sandwiches_list_chicago2.csv'
    , encoding='utf-8'
    , index_col=0
)
df.head()
```

```python
lat = []
lng = []

for idx in tqdm(df.index):
    if df['Address'][idx] != 'Multiple':
        target_name = df.loc[idx, 'Address'] + ' Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['Lat'] = lat
df['Lng'] = lng
df.head()
```

```python
map = folium.Map(
      location=[df['Lat'].mean(), df['Lng'].mean()]
    , zoom_start=11
)

for idx in df.index:
    if df['Address'][idx] != 'Multiple':
        folium.Marker(
              [df['Lat'][idx], df['Lng'][idx]]
            , popup=df['Cafe'][idx]
        ).add_to(map)
        
# folium.Marker(
#       [df['Lat'].mean(), df['Lng'].mean()]
#     , popup='Center'
# ).add_to(map)

map
```
